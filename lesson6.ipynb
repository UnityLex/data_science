```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'revenue': [4200, 5100, 4800, 6200, 7100, 6800]
}

df = pd.DataFrame(data)
print(df)
print(f"\nAverage Revenue: ${df['revenue'].mean():,.0f}")
print(f"Best Month: {df.loc[df['revenue'].idxmax(), 'month']}")
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df['month'], df['revenue'], marker='o', color='royalblue', linewidth=2)
plt.fill_between(df['month'], df['revenue'], alpha=0.1, color='royalblue')
plt.axhline(y=df['revenue'].mean(), color='red', linestyle='--', label=f"Avg: ${df['revenue'].mean():,.0f}")
plt.title('Monthly Revenue — H1', fontsize=14, fontweight='bold')
plt.ylabel('Revenue ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
